## Data Concatenation

```python
df1.set_index('Country Code', inplace=True)
df2.set_index('Country Code', inplace=True)
```

```python
df1.head()
```

```python
df2.head()
```

```python
data = pd.concat([df1,df2],axis=1)
```

## Data Cleaning

### Drop Rows

```python
# Drop rows with all null values
data.dropna(how='all',inplace=True)
```

```python
# Drop rows where 'Region' is null to remove 'Aggregates' and keep only countries
data.dropna(subset=['Region'], inplace=True)
```

```python
data
```

### Drop Columns

```python
# Drop columns with all null values
data.dropna(how='all',inplace=True, axis=1)
```

```python
# Drop unnecessary columns including duplicate columnd i.e. 'TableName' or 'Country Name'
data.drop(labels=['SpecialNotes', 'Country Name', 'Indicator Name',
                     'Indicator Code'],axis=1,inplace=True)
```

```python
data.head()
```

```python
# Assignment requires data only from year 2017 onwards,
# therefore dropping columns prior to 2017
data.drop(data.columns[3:-4], axis = 1,inplace = True)
```

```python
# Set index column name
data.index.name = 'CountryCode'
```

## Export to CSV

```python
data.to_csv('data.csv')
```
